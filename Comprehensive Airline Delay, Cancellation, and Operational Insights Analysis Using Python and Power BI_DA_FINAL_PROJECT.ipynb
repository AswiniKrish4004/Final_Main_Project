```python
# Removed duplicate records to ensure each flight is counted only once.
df.drop_duplicates(inplace=True)
df
```

```python
# Fix inconsistent text
#df['AIRLINE'] = df['AIRLINE'].str.strip().str.capitalize()
df['AIRLINE'] = df['AIRLINE'].astype(str).str.strip().str.title()
df
```

```python

#Removed a redundant column to simplify the dataset.
df = df.drop('AIRLINE_DOT', axis=1)


df
```

```python
#Removed leading and trailing spaces from airport and city names.
df['ORIGIN_CITY'] = df['ORIGIN_CITY'].str.strip()
df['ORIGIN'] = df['ORIGIN'].str.strip()
df['DEST'] = df['DEST'].str.strip()
df['DEST_CITY'] = df['DEST_CITY'].str.strip()

df
```

```python
#Replaced missing cancellation codes for non-cancelled flights with "Not Cancelled".
df.loc[df['CANCELLED'] == 0, 'CANCELLATION_CODE'] = (df.loc[df['CANCELLED'] == 0, 'CANCELLATION_CODE'].fillna('Not Cancelled'))

df['CANCELLATION_CODE']
```

```python
#Set all operational flight-related values to 0 for cancelled flights because the flight was not operated.

ops_cols = ['DEP_TIME', 'ARR_TIME',
            'WHEELS_OFF', 'WHEELS_ON',
            'AIR_TIME', 'TAXI_OUT', 'TAXI_IN','DEP_DELAY', 'ARR_DELAY','ELAPSED_TIME']

df.loc[df['CANCELLED'] == 1, ops_cols] = 0
df
```

```python
#Filtered cancelled flights and displayed the operational columns to verify that their values were correctly set to 0.
mask = df['CANCELLED'] == 1

print(df.loc[mask, ops_cols].head())
```

```python
#Converted the flight date column from text format to datetime format for date-based analysis.

import pandas as pd
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], dayfirst=True)
```

```python
#Extracted the year from the flight date to analyze yearly trends and patterns.
df['YEAR'] = df['FL_DATE'].dt.year
```

```python
df.dtypes
```

```python

#Standardized departure and arrival time columns by converting them into a consistent HH:MM format.
time_cols = ['CRS_DEP_TIME','DEP_TIME','CRS_ARR_TIME','ARR_TIME']

for col in time_cols:
    df[col + '_TIME'] = pd.to_datetime(
        df[col].astype('Int64').astype(str).str.zfill(4),
        format='%H%M',
        errors='coerce'
    ).dt.strftime('%H:%M')
```

```python
#Extracted the Month and Month Name  from the flight date to analyze Monthly trends and patterns.
df['MONTH'] = df['FL_DATE'].dt.month
df['MONTH_NAME'] = df['FL_DATE'].dt.month_name()
```

```python
#Imputed missing values in diverted flights by replacing them with the median value from the same airline and year to maintain data consistency and reduce bias.
cols = ['WHEELS_ON', 'ARR_DELAY', 'AIR_TIME', 'TAXI_IN', 'ELAPSED_TIME','ARR_TIME']

mask1 = (df['CANCELLED'] == 0) & (df['DIVERTED'] == 1)

for col in cols:
    df.loc[mask1, col] = (
        df.loc[mask1]
          .groupby(['AIRLINE', 'YEAR'])[col]
          .transform(lambda x: x.fillna(x.median()))
    )
```

```python
df.loc[mask1, cols].isnull().sum()
```

```python
#Imputed missing values by replacing them with the median of the corresponding airline, ensuring more accurate and consistent data within each airline category.
for col in cols:
    df[col] = df.groupby('AIRLINE')[col].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
#Set all delay reason columns to 0 for flights that arrived on time or early, since no arrival delay occurred.
delay_cols = ['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']
df.loc[df['ARR_DELAY'] <= 0, delay_cols] = 0
```

```python
#Calculated the skewness of numerical columns to understand the distribution shape and identify potential outliers.
num_cols = [
    'DEP_DELAY',
    'ARR_DELAY',
    'AIR_TIME',
    'DISTANCE',
    'TAXI_OUT',
    'TAXI_IN',
    'ELAPSED_TIME'
]

df[num_cols].skew()
```

```python
#Created a copy of the cleaned dataset to preserve the original data before applying further transformations or outlier treatment.

df_clean = df.copy()
```

```python
#Removed extreme outliers from numerical columns using the IQR method and compared the dataset size before and after cleaning.

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.8 * IQR
    upper = Q3 + 1.8 * IQR

    df = df[
        (df[col] >= lower) &
        (df[col] <= upper)
    ]

print("Original Shape:", df_clean.shape)
print("New Shape:", df.shape)
print("Rows Removed:", len(df_clean) - len(df))
```

```python
#Removed extreme outliers from numerical columns using the IQR method

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.8 * IQR
    upper = Q3 + 1.8 * IQR

    count = ((df[col] < lower) | (df[col] > upper)).sum()

    print(f"{col}: {count}")
```

```python
print(f"Original rows: {len(df_clean)}")
print(f"Remaining rows: {len(df)}")
print(f"Removed rows: {len(df_clean) - len(df)}")
print(f"Percentage removed: {((len(df_clean)-len(df))/len(df_clean))*100:.2f}%")
```

##**Chart 3 :Flight Cancellation Distribution (Pie)**
